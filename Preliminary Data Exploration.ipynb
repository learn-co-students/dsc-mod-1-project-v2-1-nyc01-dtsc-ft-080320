```python
minbudget = dfdi["tn.movie_budgets.csv"].production_budget.min()
maxbudget = dfdi["tn.movie_budgets.csv"].production_budget.max()
```

```python
dfdi["imdb.title.basics.csv"].head()
base_genres = dfdi["imdb.title.basics.csv"].dropna(axis=0,subset=["genres"])
```

```python
genres = base_genres["genres"].map(lambda x: x.split(","))
#type(base_genres["genres"][0])
```

```python
genres
```

```python
new_genres = {}
for i in genres.index:
    for genre in genres[i]:
        try:
            new_genres[genre].append(i)
        except:
            new_genres[genre] = [i]
print(new_genres.keys())
```

```python
likeys = list(new_genres.keys())
```

```python
keydf = pd.DataFrame(0, index=np.arange(len(genres)), columns = likeys)
```

```python
mergedgenres = base_genres.join(keydf)
```

```python
for key,value in new_genres.items():
    mergedgenres.loc[value,key] = 1
```
